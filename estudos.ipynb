```python
import pandas as pd # quando ussamos o 'pd' é igual ao eles do mysql
caminho = r'C:\Users\leone\OneDrive\Desktop\ana\pandas\vendas.csv'
tb = pd.read_csv(caminho)

tb.head(1) # sem passar parametro é as 5 primeiras linhas
```

```python
# aonde fariamos um SELECT SUM('VAOLOR_VENDA') FROM VENDAS | NO PANDAS ACESSAMOS A COLUNA COM .SUM()
faturamento = tb['valor_venda'].sum()
print(f"O faturamento foi de R$: {faturamento:,.2f}")
```

```python
# aonde apos fazer o SELECT SUM()... E UM GROUP BY categoria_produto, aqui vamos declarar antes da soma da coluna o que desejamos fazer
faturaentoPorCategoria = tb.groupby('categoria_produto')['valor_venda'].sum()
print(f"{faturaentoPorCategoria}")

# a parte apos o .sum() é para organização de valores, aonde  vai pegar o resultado da soma e organizar | o parametro passado é para saber se é ordem degrecente ou crescente, menor para o maior (asceding=True) colocando False ele inverte. Olhando para o SQL seria como o nosso Order BY SUM(VALOR_VENDA) DESC
faturaentoPorCategoria = tb.groupby('categoria_produto')['valor_venda'].sum().sort_values(ascending=False)
print(f"{faturaentoPorCategoria}")
```

```python
#o value_counts é como o nosso count()
rankingVendasQtd = tb['representante'].value_counts()
rankingVendasValor = tb.groupby('representante')['valor_venda'].sum()
print(f"{rankingVendasQtd} f{rankingVendasValor}")

#caso queria criar algum filtro como por exemplo varremos a coluna de represntantes aonde pegamos somente os valores diferentes de 'Ana' (ele retorna a cada linha um valor de True ou False = Boolean Mask) é como o nosso WHERE REPRESENTANTE <> 'Ana'
rankinfVendasQtdFiltrada = tb[tb['representante'] != 'Ana']['representante'].value_counts()
print(rankinfVendasQtdFiltrada)
```

```python
#Convertendo a coluna para data real
tb['data_venda'] = pd.to_datetime(tb['data_venda'])
print(tb.data_venda.dtypes)
```

```python
#1- Qual a região que teve o maior faturamento total?
analiseRegional = tb.groupby('regiao')['valor_venda'].sum()
print(analiseRegional)

#2- Como você faria para criar um novo DataFrame chamado vendas_notebook contendo apenas as linhas onde o produto é 'Notebook'?
vendas_notebook = tb[tb['produto'] == 'Notebook']
print(vendas_notebook)

#3- Qual região gerou mais faturamento APENAS com a venda de Notebooks?
analiseRegionalFiltradaNotebooks = vendas_notebook.groupby('regiao')['valor_venda'].sum().sort_values(ascending=False)
print(analiseRegionalFiltradaNotebooks)
```

```python
# Criar novas colunas na nossa base bruta (tb), no mysql fazemos da seguinte maneira SELECT REPRESENTANTE,VALOR_VENDA,(VALOR_VENDA * 0.010)B AS COMISSAO FROM TB; | aqui vamos declarar a coluna nova como se estivessemos declarando uma variavel porem no modelo de tabela['colunaNova'] = calculo ou valor da coluna novo 
tb['comissao'] = tb['valor_venda'] * 0.010
tb.head()

comissaoPorRepresentante = tb.groupby('representante')['comissao'].sum().sort_values(ascending=False)
print(comissaoPorRepresentante)
```
