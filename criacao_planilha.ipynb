```python
dados = [
    {
        "CIDADES_ROTA": "MORUNGABA",
        "ROTA": "MINAS",
        "DIA_FATURAMENTO1": "TERÇA-FEIRA",
        "DIA_FATURAMENTO2": "NAO POSSUI",
        "SETOR": "SP",
        "TEMPO_ROTA": "1D",
        "DIA_FATURAMENTO3": "NAO POSSUI"
    },
    {
        "CIDADES_ROTA": "SENADOR AMARAL",
        "ROTA": "MINAS",
        "DIA_FATURAMENTO1": "TERÇA-FEIRA",
        "DIA_FATURAMENTO2": "NAO POSSUI",
        "SETOR": "MG",
        "TEMPO_ROTA": "1D\r",
        "DIA_FATURAMENTO3": "NAO POSSUI"
    },
    {
        "CIDADES_ROTA": "BORDA DA MATA",
        "ROTA": "MINAS",
        "DIA_FATURAMENTO1": "TERÇA-FEIRA",
        "DIA_FATURAMENTO2": "NAO POSSUI",
        "SETOR": "MG",
        "TEMPO_ROTA": "1D\r",
        "DIA_FATURAMENTO3": "NAO POSSUI"
    }
]
```

```python
import pandas as pd  
from datetime import datetime 
from openpyxl.styles import PatternFill, Font, Alignment  

tb_dados = pd.DataFrame(dados)
data_hoje = datetime.now().strftime('%d/%m/%y')
nm_arquivo = "grade_faturamento.xlsx"

with pd.ExcelWriter(nm_arquivo, engine='openpyxl') as cursor: # pandas com execel editor e o openpyxl
    tb_dados.to_excel(cursor, index=False, sheet_name='Grade', startrow=1) #abrir o execel com a tb-dados *com* o cursor(...) o index=False = não numerar as linhas, startrow=1 = pular a linha 0(1) e começar na 1(2)
    planilha_aba= cursor.sheets['Grade']
    total_colunas = len(tb_dados.columns)

    planilha_aba.merge_cells(start_row =1, start_column =1, end_row = 1, end_column = total_colunas)# mesclando o topo
    topo_txt_centralizado = Alignment(horizontal='center', vertical='center')
    topo_txt_fonte_branca = Font(color='f0f0f0', bold=True, size=14)
    topo_fundo = PatternFill(start_color='181848', end_color='181848', fill_type='solid')

    topo_txt = planilha_aba.cell(row=1,column=1)
    topo_txt.value = f"Grade de Atendimento | Emitido em: {data_hoje}"
    topo_txt.fill = topo_fundo
    topo_txt.font = topo_txt_fonte_branca
    topo_txt.alignment = topo_txt_centralizado
    planilha_aba.row_dimensions[1].height=40
```
