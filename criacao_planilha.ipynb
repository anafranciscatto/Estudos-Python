```python
dados = [
    {
        "CIDADES_ROTA": "MORUNGABA",
        "ROTA": "MINAS",
        "DIA_FATURAMENTO1": "TERÇA-FEIRA",
        "DIA_FATURAMENTO2": "NAO POSSUI",
        "SETOR": "SP",
        "TEMPO_ROTA": "1D",
        "DIA_FATURAMENTO3": "NAO POSSUI"
    },
    {
        "CIDADES_ROTA": "SENADOR AMARAL",
        "ROTA": "MINAS",
        "DIA_FATURAMENTO1": "TERÇA-FEIRA",
        "DIA_FATURAMENTO2": "NAO POSSUI",
        "SETOR": "MG",
        "TEMPO_ROTA": "1D\r",
        "DIA_FATURAMENTO3": "NAO POSSUI"
    },
    {
        "CIDADES_ROTA": "BORDA DA MATA",
        "ROTA": "MINAS",
        "DIA_FATURAMENTO1": "TERÇA-FEIRA",
        "DIA_FATURAMENTO2": "NAO POSSUI",
        "SETOR": "MG",
        "TEMPO_ROTA": "1D\r",
        "DIA_FATURAMENTO3": "NAO POSSUI"
    }
]
```

```python
import pandas as pd  
import io  
from datetime import datetime 
from openpyxl.styles import PatternFill, Font, Alignment  

tabela_dados = pd.DataFrame(dados)
data_atual = datetime.now().strftime('%d/%m/%Y')

# 3. Cria um "arquivo virtual" na memória para não precisar salvar no disco do servidor
arquivo_em_memoria = io.BytesIO()

# 4. Inicia a escrita do arquivo Excel usando o motor 'openpyxl'
with pd.ExcelWriter(arquivo_em_memoria, engine='openpyxl') as escritor:
    
    # Escreve a tabela na aba 'Grade', começando na linha 2 (para sobrar espaço pro título)
    tabela_dados.to_excel(escritor, index=False, sheet_name='Grade', startrow=1)
    
    # Acessa os controles internos da planilha e da aba específica
    planilha = escritor.book
    aba = escritor.sheets['Grade']
    
    # Descobre quantas colunas a tabela tem para saber até onde mesclar o título
    total_colunas = len(tabela_dados.columns)
    
    # Mescla as células da primeira linha (da coluna 1 até a última coluna da tabela)
    aba.merge_cells(start_row=1, start_column=1, end_row=1, end_column=total_colunas)
    
    # Seleciona a célula do título (A1) e define o texto com a data de emissão
    celula_principal = aba.cell(row=1, column=1)
    celula_principal.value = f"GRADE DE ATENDIMENTO | Emitido em: {data_atual}"
    
    # Define a cor de fundo: Azul Escuro Atento (HEX: 181848)
    preenchimento_azul = PatternFill(start_color='181848', end_color='181848', fill_type='solid')
    
    # Define o estilo da letra: Cor Branca, Negrito e Tamanho 14
    estilo_fonte_branca = Font(color='FFFFFF', bold=True, size=14)
    
    # Define o alinhamento: Texto centralizado tanto na largura quanto na altura
    centralizar_texto = Alignment(horizontal='center', vertical='center')
    
    # Aplica as cores, fontes e alinhamento na célula do título
    celula_principal.fill = preenchimento_azul
    celula_principal.font = estilo_fonte_branca
    celula_principal.alignment = centralizar_texto
    
    # Ajusta a altura da primeira linha (título) para 35 pixels para ficar mais elegante
    aba.row_dimensions[1].height = 35

# 5. Salva o conteúdo da memória em um arquivo físico no computador (para seu teste)
with open("Faturamento_Grade2.xlsx", "wb") as arquivo_final:
    arquivo_final.write(arquivo_em_memoria.getvalue())

# Mensagem de confirmação no console
print("arquivo gerado!")
```
